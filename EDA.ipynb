## 2. Team Analysis:

### - What is the win percentage of each team?

```python
Team_total_wins = df.groupby(["winner"], as_index=False)["winner"].value_counts().sort_values("count", ascending=False)
Team_total_wins["percent"] = Team_total_wins["count"]*100/Total_matches
Team_total_wins
```

#### From the above graph we can see that Mumabi Indians have highest win percentage.

### - How does the performance of teams vary by venue?

```python
pivot_table = venue_performance.pivot(index="venue", columns="winner", values="count").fillna(0)
pivot_table
```

# EDA

## 1. Batting Performance:

### - Which players have the highest total runs?

```python
data.columns
```

```python
highest_runs = data.groupby("batter", as_index=False)["batsman_runs"].sum().sort_values("batsman_runs", ascending=False).head(10)
highest_runs
```

```python
plt.figure(figsize=(12,6))
ax = sns.barplot(x="batter", y="batsman_runs", data=highest_runs)
for bar in ax.containers:
    ax.bar_label(bar)
plt.xlabel("Players")
plt.ylabel("Total IPL Runs")
plt.show()
```

#### From the above graph we can see that Virat Kohli has the highest total runs (8014).

###  Who are the most consistent batsmen (highest average)?

```python
batsman_stats = data.groupby(['batter']).agg(
    total_runs=('batsman_runs', 'sum'),
    dismissals=('is_wicket', 'sum')
).reset_index()

batsman_stats['average'] = round(batsman_stats['total_runs'] / batsman_stats['dismissals'].replace(0, 1),2)
batsman_stats
most_consistent_batsmen = batsman_stats.sort_values("average", ascending=False).head(10)
most_consistent_batsmen
```

```python
plt.figure(figsize=(15,6))
ax = sns.barplot(x="batter", y="average", data = most_consistent_batsmen, palette="hsv")
for bar in ax.containers:
    ax.bar_label(bar)
plt.title("Most Consistent Batsmen", fontsize=15)
plt.xlabel("Players",fontsize=12)
plt.ylabel("Average Strike Rate",fontsize=12)
plt.show()
```

#### From the above graph we can see that "Iqbal Abdulla" has the highest avergae.

### - What is the distribution of runs scored in boundaries (4s and 6s) versus singles?

```python
runs = data["batsman_runs"].value_counts().reset_index().sort_values("batsman_runs", ascending=False)
runs.columns = ['Run Type', 'Count']
runs
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Run Type', y='Count', data=runs, palette='viridis')
plt.title('Distribution of Runs Scored: Boundaries vs Singles')
plt.xlabel('Run Type')
plt.ylabel('Count')
plt.show()
```

## 2. Bowling Performance:

### - Which bowlers have taken the most wickets?

```python
highest_wickets = data.groupby("bowler", as_index=False)["is_wicket"].sum().sort_values("is_wicket", ascending=False).head(10)
highest_wickets.columns = ["Bowler","Total Wickets"]
highest_wickets
```

```python
plt.figure(figsize=(12,6))
ax = sns.barplot(x="Bowler", y="Total Wickets", data=highest_wickets)
for bar in ax.containers:
    ax.bar_label(bar)
plt.xlabel("Players")
plt.ylabel("Total IPL Wickets")
plt.show()
```

#### From the above graph we can see that "YS Chahal" has taken the most wickets.

### - Who are the most economical bowlers?

```python
# grouping bowler data 
bowler_stats = data.groupby(["bowler"]).agg(
    total_run_conceded = ("total_runs","sum"),
    total_balls = ("ball", "count")
).reset_index()
# converting balls into overs
bowler_stats["over bowled"] = round(bowler_stats["total_balls"]/6,1)

# calculating economy
bowler_stats["economy"] = round(bowler_stats["total_run_conceded"]/bowler_stats["over bowled"],2)

top_economical_bowlers = bowler_stats.sort_values("economy", ascending=False)
top_economical_bowlers
```

```python
# Filter out bowlers with very few overs bowled to ensure meaningful analysis
bowler_stats_filtered = bowler_stats[bowler_stats['over bowled'] >= 10]

# Sort bowlers by economy rate in ascending order
most_economical_bowlers = bowler_stats_filtered.sort_values(by='economy', ascending=True).head(10)

most_economical_bowlers
```

```python
plt.figure(figsize= (15,6))
ax = sns.barplot(x="bowler", y="economy", data=most_economical_bowlers)
for bar in ax.containers:
    ax.bar_label(bar)
plt.title("Most Economical Bowlers", fontsize=15)
plt.xlabel("Bowlers",fontsize=12)
plt.ylabel("Economy",fontsize=12)
plt.show()
```

#### From the above graph we can see that "Sohail Tanvir" has the best economy.

## 3. Extras:

### - What are the most common types of extras (no balls, wides, etc.)

```python
extra_type = data.groupby("extras_type", as_index=False).agg(
    count=("extras_type", "count"),
    total_extra_runs=("extra_runs", "sum")
).sort_values("count")
extra_type
```

```python
ax = sns.barplot(x="extras_type", y="count", data=extra_type, palette="coolwarm")
for bar in ax.containers:
    ax.bar_label(bar)
plt.xlabel("Extra Types",fontsize=11)
plt.ylabel("Count",fontsize=11)
plt.show()
```

### - How often do extras significantly contribute to the total score?

```python
ax = sns.barplot(x="extras_type", y="total_extra_runs", data=extra_type, palette = "cool")
for bar in ax.containers:
    ax.bar_label(bar)
plt.xlabel("Extra Types",fontsize=11)
plt.ylabel("Contribution in runs",fontsize=11)
```
